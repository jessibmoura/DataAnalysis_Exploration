```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
consumo = pd.read_excel('files/consumo.xlsx')
consumo
```

```python
consumo.iloc[:, 1:] = consumo.iloc[:, 1:] * 100
```

#### Gráfico de barras

##### 15 anos

```python
# Transformar o DataFrame para o formato longo e acessar os valores de todos os paises
cores = {
    'Russia': 'darkred',
    'Haiti': 'green',
    'USA': 'blue',
    'China': 'red',
    'Uruguai': 'darkcyan',
    'Paraguai': 'orange'
}

consumo_long = pd.melt(consumo, id_vars=['País'], var_name='Ano', value_name='Percentual')

plt.figure(figsize=(20, 5))

grafico = sns.barplot(data=consumo_long, x='Ano', y='Percentual', hue='País', palette=cores)

sns.set_theme()

plt.title('Razão Consumo x Importação de vinho')
plt.xlabel('Ano')
plt.ylim(0, 120)
plt.ylabel('Percentual de Consumo (%)')

plt.legend(title='País', bbox_to_anchor=(0.5, -0.15), loc='upper center', ncol=3)

# plt.tight_layout()
plt.show()
```

```python
cores = {
    'Russia': 'darkred',
    'Haiti': 'green',
    'USA': 'blue',
    'China': 'red',
    'Uruguai': 'darkcyan',
    'Paraguai': 'orange'
}

consumo_long = pd.melt(consumo, id_vars=['País'], var_name='Ano', value_name='Percentual')

plt.figure(figsize=(20, 5))

grafico = sns.barplot(data=consumo_long, x='Ano', y='Percentual', hue='País', palette=cores)

# Adicionar título e rótulos aos eixos
plt.title('Razão Consumo x Importação de vinho')
plt.xlabel('Ano')
plt.ylim(0, 120)
plt.ylabel('Percentual de Consumo (%)')

plt.legend(title='País', bbox_to_anchor=(0.5, -0.15), loc='upper center', ncol=3)

sns.set_theme()
# remove a linha de cima pra ver melhor os valores
grafico.spines['top'].set_visible(False)
grafico.spines['right'].set_visible(False)

# Adicionar os valores nas barras
for p in grafico.patches:
    grafico.annotate(format(p.get_height(), '.1f'),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha = 'center', va = 'center',
                     xytext = (0, 10), 
                     textcoords = 'offset points')


plt.tight_layout()
plt.show()
```

```python
# Transformar o DataFrame para o formato longo
consumo_long = pd.melt(consumo, id_vars=['País'], var_name='Ano', value_name='Percentual')

# Definir o dicionário de cores
cores = {
    'Russia': 'darkred',
    'Haiti': 'green',
    'USA': 'blue',
    'China': 'red',
    'Uruguai': 'darkcyan',
    'Paraguai': 'orange'
}

# Criar o gráfico
plt.figure(figsize=(10, 6))

grafico = sns.pointplot(data=consumo_long, x='Ano', y='Percentual', hue='País', palette=cores, marker='o')

# Adicionar título e rótulos aos eixos
plt.title('Razão Consumo x Importação de vinho')
plt.xlabel('Ano')
plt.ylabel('Percentual de Consumo (%)')
plt.ylim(0, 120)
plt.legend(title='País', bbox_to_anchor=(0.5, -0.15), loc='upper center', ncol=3)

plt.tight_layout()
plt.show()
```
