```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import table
```

```python
prod_exp_br = pd.read_excel("files/prod_e_export_brazil_15_anos.xlsx",engine="openpyxl",index_col=None)
```

```python
prod_exp_br
```

```python
exp_br = prod_exp_br[prod_exp_br['Variable'] == 'Exports']
exp_br
```

```python
prod_br = prod_exp_br[prod_exp_br['Variable'] == 'Production']
prod_br
```

```python
def convert_value(x):
    y = x*1000
    return (f'{y} hl')

temp_br = prod_br['Quantity']
values = temp_br.apply(convert_value)
prod_br['Quantity (hl)'] = values

temp_br = exp_br['Quantity']
values = temp_br.apply(convert_value)
exp_br['Quantity (hl)'] = values
```

```python
exp_br
```

```python
prod_br
```

```python
prod_br.rename(columns={'Region/Country':'País','Year':'Ano','Quantity (hl)':'Quantidade (hl)'},inplace=True)
exp_br.rename(columns={'Region/Country':'País','Year':'Ano','Quantity (hl)':'Quantidade (hl)'},inplace=True)
```

```python
prod_br.columns
```

```python
prod_br_final = prod_br[['Ano','País','Quantidade (hl)']]
prod_br_final
```

```python
exp_br_final = exp_br[['Ano','País','Quantidade (hl)']]
exp_br_final
```

```python
prod_br_final_5 = prod_br_final[prod_br_final['Ano'] >= 2018]
prod_br_final_5
```

```python
exp_br_final_5 = exp_br_final[exp_br_final['Ano'] >= 2018]
exp_br_final_5
```

```python
fig, ax = plt.subplots(figsize=(10, 3))  # Ajuste o tamanho conforme necessário

# Remover os eixos
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
ax.set_frame_on(False)
ax.set_title('Produção de Vinho do Brasil (2018 - 2022)', fontsize=14, weight='bold')

# Criar a tabela e adicionar ao gráfico
tbl = table(ax, prod_br_final_5, loc='center', cellLoc='center', colWidths=[0.1]*len(prod_br_final_5.columns))

# Estilizar a tabela
tbl.auto_set_font_size(False)
# tbl.auto_set_column_width([0])
tbl.set_fontsize(11)
tbl.scale(2, 2)
fig.text(0.5, -0.1, 'Média da produção (2018-2022): 2,729,000 hl', ha='center', fontsize=13, style='normal')
fig.text(0.5, -0.2, 'Figura 1: Produção de vinho no Brasil', ha='center', fontsize=11, style='italic')



# Salvar a tabela como imagem
plt.savefig('Output files/tabela_producao_br_5.png', bbox_inches='tight', dpi=300)

# Exibir a imagem
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 3))  # Ajuste o tamanho conforme necessário

# Remover os eixos
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
ax.set_frame_on(False)
ax.set_title('Exportação de Vinho do Brasil (2018 - 2022)', fontsize=14, weight='bold')

# Criar a tabela e adicionar ao gráfico
tbl = table(ax, exp_br_final_5, loc='center', cellLoc='center', colWidths=[0.1]*len(exp_br_final_5.columns))

# Estilizar a tabela
tbl.auto_set_font_size(False)
# tbl.auto_set_column_width([0])
tbl.set_fontsize(11)
tbl.scale(2, 2)
fig.text(0.5, -0.1, 'Média da produção (2018-2022): 57,600 hl', ha='center', fontsize=13, style='normal')
fig.text(0.5, -0.2, 'Figura 2: Exportação de vinho no Brasil', ha='center', fontsize=11, style='italic')



# Salvar a tabela como imagem
plt.savefig('Output files/tabela_export_br_5.png', bbox_inches='tight', dpi=300)

# Exibir a imagem
plt.show()
```
