##### Create new dataframe

```python
'''
    Auxiliary functions
'''

def _getTotalExporters(liters_table,year):
    temp = liters_table[year].map(lambda x:1 if x != 0 else 0)
    totalExporters = temp.sum()
    # print(f'Total Exporters: {totalExporters}')
    return totalExporters
    
def _getTotalProfit(dolars_table,year):
    totalSum = dolars_table[year].sum()
    # print(f'Total Profit ($): {totalSum}')
    return totalSum

def _getPercentage(dolars,total):
    per = round(dolars / total, 2)
    # print(f'Percentage of total (%): {per}')
    return per
    
```

```python
def _getCountryMostLitersInfo(liters_table,dolars_table,year):
    '''
    year: year its analysing
    ml_country: country that has exported most liters of wine
    ml_liters: how much wine has been exported (liters)
    ml_dol: how much has been the country's profit
    ml_per: percentage of the country's profit by total profit
    '''
    
    id = liters_table[year].idxmax() # get index of the country that exported the most
    ml_liters = liters_table[year][id] # gets the highest value
    ml_country = liters_table['País'][id] #gets the corresponding country
    ml_dol = dolars_table[year][id]
    
    # print("---------------------------------------")
    # print(f'Year: {year}')
    # print(f'Most Liters Country: {ml_country}')
    # print(f'Most Liters Quantity (Kg): {ml_quant:,}')
    # print(f'Most Liters Value ($): {ml_dol:,}')
    tot = _getTotalProfit(dolars_table,year)
    ml_per = _getPercentage(ml_dol,tot)
    
    return ml_country, ml_liters, ml_dol, ml_per

def _getCountryMostProfit(liters_table,dolars_table,year):
    '''
    year: year its analysing
    mp_country: country that has generated most profit
    mp_liters: how much wine has been exported (liters)
    mp_dol: how much has been the country's profit
    mp_per: percentage of the country's profit by total profit
    '''
    
    id = dolars_table[year].idxmax() # get index of the country that generated more profit
    mp_dol = dolars_table[year][id] # gets the highest value
    mp_liters = liters_table[year][id]
    mp_country = dolars_table['País'][id] #gets the corresponding country
    
    # print(f'Most Profit Country: {country}')
    # print(f'Most Profit Quantity (Kg): {liters:,}')
    # print(f'Most Profit Value ($): {dolars:,}')
    tot = _getTotalProfit(dolars_table,year)
    mp_per = _getPercentage(mp_dol,tot)
    
    return mp_country, mp_liters, mp_dol, mp_per
```

```python
COLUMN_NAMES = [
                'Year',
                'Most Liters Country',
                'Most Liters Quantity (Kg)',
                'Most Liters Value ($)',
                'ML Percentage of total (%)',
                'Most Profit Country',
                'Most Profit Quantity (Kg)',
                'Most Profit Value ($)',
                'MP Percentage of total (%)',
                'Total Exporters',
                'Total Value ($)'
                ]

analysis_df = pd.DataFrame(columns=COLUMN_NAMES)
```

```python
temporaryDict = {
    'Year': '',
    'Most Liters Country': '',
    'Most Liters Quantity (Kg)': '',
    'Most Liters Value ($)': '',
    'ML Percentage of total (%)': '',
    'Most Profit Country': '',
    'Most Profit Quantity (Kg)': '',
    'Most Profit Value ($)': '',
    'MP Percentage of total (%)': '',
    'Total Exporters': '',
    'Total Value ($)': ''
}
```

```python
for col in qtde_litros.columns:
    
    if 'País' not in col:
        
        year = col
        
        ml_country, ml_liters, ml_dol, ml_per = _getCountryMostLitersInfo(qtde_litros, qtde_dolares, year)
        mp_country, mp_liters, mp_dol, mp_per = _getCountryMostProfit(qtde_litros, qtde_dolares, year)
        totalExporters = _getTotalExporters(qtde_litros,year)
        totalProfit = _getTotalProfit(qtde_dolares, year)
        
        temporaryDict['Year'] = year
        temporaryDict['Most Liters Country'] = ml_country
        temporaryDict['Most Liters Quantity (Kg)'] = ml_liters
        temporaryDict['Most Liters Value ($)'] = ml_dol
        temporaryDict['ML Percentage of total (%)'] = ml_per
        temporaryDict['Most Profit Country'] = mp_country
        temporaryDict['Most Profit Quantity (Kg)'] = mp_liters
        temporaryDict['Most Profit Value ($)'] = mp_dol
        temporaryDict['MP Percentage of total (%)'] = mp_per
        temporaryDict['Total Exporters'] = totalExporters
        temporaryDict['Total Value ($)'] = totalProfit
        
        analysis_df = analysis_df._append(temporaryDict, ignore_index=True)
        
        
        
```
