### Imports

```python
import pandas as pd
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
import numpy as np
```

```python
pd.options.display.float_format = "{:.2f}".format # change the way the mean is displayed
```

### Data Analysis

```python
dataset = pd.read_csv('bases/A150850189_28_143_208.csv', encoding='iso-8859-1',
                      skiprows=3, skipfooter=12, sep=";", thousands=".", decimal=",")
```

```python
dataset.head() # returns the first 5 rows
```

```python
dataset.mean(numeric_only=True) # ignores all non numeric 
```

```python
dataset.info(verbose=False)
```

```python
dataset['2008/Ago']
```

```python
axis = dataset.plot(
    x="Unidade da Federação",
    y="2008/Ago",
    kind="bar"
)

axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
axis.tick_params(axis='x', rotation=45)
# plt.xticks(dataset["Unidade da Federação"], dataset["2008/Ago"], rotation=45)
plt.title("Valor por unidade da federação")

plt.show()
```
